```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# EDA for understand data
# 1) Descriptive statistics
# 2) Data Visualization- plots and charts. Histogram, box plot, density curve
# 3) Data Aggregation/grouping - groupby() 

# Inferential Statistics - Hypothesis Testing -  Comparisions of Means or Averages of Groups/levels/Classes
# Fundamental Assumption of hypothsis testing is means or Averages of Groups/levels must be different. 
# groupby() is very imp in hypotheis.

# Fundamental Assumption is Numerical continuous Variables must be closer to normal distribution

# The procedure for conducting the hypothesis testing is you shourld do
# 1) groupby() -  numeric variables & Categorical variables  or crosstab() - (Both Variable are non numeirc and categorical)

# 2) groupby() will confirm mean difference and also identify the number of levesl/groups/classes in categorical variable

# 3) Identify the Hypotheis Test to be conducted

# One variable Numeric and Other Varialbe Categorical with exactly 2 levels/Groups/Classes - 2 Sample Independent T test

# One variable Numeric and Other Varialbe Categorical with more than 2 levels/Groups/Classes-Anova Single Factor or Oneway Anova

# Both variables are Categorical and Non Numeric - Chisquare Test of Independence

# 4) Frame Null Hypothesis and ALternate Hypothesis.

# 5) Split data according to Groups/Levels/Classes

# 6) Conduct the relevant hypothesis test

# 7) Interpretation of output
# p-value or probability value is less than 0.05, WE WILL REJECT NULL HYPOTHESIS
# p-value or probability value is greater than 0.05 , FAIL TO REJECT NULL(ACCEPT NULL)

# 0.05 means 5% Alpha or probable error rate and 95% Confidence Level
# other p-value like 0.01 or 0.001 used in pharma & Clinical Trials
```

```python
# test Null Average cnt of bicycles on holiday 0/1 equal.
df = pd.read_csv('D:\Datasets\day.csv')
```

```python
df.head()
```

```python
df.cnt.groupby(df.holiday).mean() # Means are different Exactly 2 Classes/Levels/groups - 2 Sample Independent T test
```

```python
# 2 Types of Sample Independent T test - groupby() -var
# 2 Sample Independent T test assuming Unequal Variance
# 2 Sample Independent T test assuming equal Variance

df.cnt.groupby(df.holiday).var()
```

```python
# Frame the Null & Alternate Hypothesis
# Null _  There is no significant differnce in Average cnt of holiday 0/1
# it means Treat Both Means or Averages as equal.(Ignore the difference)

# Alt  _  There is significant differnce in Average cnt of holiday 0/1
# it means Treat Both Means or Averages as not equal. (cannot ignore the difference)

# Is the difference in mean or Averages of groups/levels/classes is statistically significant or not
```

```python
# SPLIT DATA

holidayyes = df[df.holiday==1]
holidayno = df[df.holiday==0]
```

```python
holidayyes
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(holidayyes.cnt,holidayno.cnt, equal_var=False)

# Since pvalue=0.10299984047957059 is greater than 0.05, Fail to Reject Null. Difference in means in NOT STATISTICALLY SIGNIFICANT

# Reject Null -  Significant
# Fail to Reject Null - Not Significant
```

```python
# Means of holidayyes & holidayno

print(np.mean(holidayno.cnt))
print(np.mean(holidayyes.cnt))
```

```python
# Variance of holidayyes & holidayno
print(np.var(holidayyes.cnt))
print(np.var(holidayno.cnt))
```
