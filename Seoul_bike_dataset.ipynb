```python
# c) Histogram, Boxplot and density curve of Rented Bike count (y)

plt.figure(figsize=(10,7))
fig,ax = plt.subplots(3,1)
sns.boxplot(data['Rented Bike Count'],orient='h',ax=ax[1])
sns.kdeplot(data['Rented Bike Count'],ax=ax[0])
sns.histplot(data['Rented Bike Count'],ax=ax[2])
```

```python
# e) Bar Plot of Average Rented Bike count hourwise

plt.figure(figsize=(17,7))
ax = data['Rented Bike Count'].groupby(data.Hour).mean().plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i,color='red')
```

```python
# f) Bar Plot of Average Rented Bike count  Season Wise

ax = data['Rented Bike Count'].groupby(data.Seasons).mean().plot(kind='bar',color='green')
for i in ax.containers:
    ax.bar_label(i,color='red')
```

```python
# g) Bar Plot of Average Rented Bike count  holiday wise

ax = data['Rented Bike Count'].groupby(data.Holiday).mean().plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i,color='red')
```

```python
# h) Bar Plot of Average Rented Bike count  Functional Wise

ax = data['Rented Bike Count'].groupby(data['Functioning Day']).mean().plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i,color = 'orange')
```

```python
# i) Stacked BAr Plot and Cross tab of Season and Holiday
ax = pd.crosstab(data.Seasons,data.Holiday).plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i)
```

```python
data = data.drop(['Dew point temperature(°C)'],axis = 1)
```

```python
data.shape
```

```python
# k) Test Null Average Rented Bike count of Holiday equal
data['Rented Bike Count'].groupby(data.Holiday).mean()
```

```python
from scipy.stats import ttest_ind
```

```python
holdyes = data[data.Holiday == 'Holiday']
holdno  = data[data.Holiday == 'No Holiday']
```

```python
ttest_ind(holdyes['Rented Bike Count'],holdno['Rented Bike Count'])

# Since p-value 1.2156443235024855e-11 is less than 0.05 reject null 
```

```python
# l) Test Null Average Visibility of Holiday equal
ttest_ind(holdyes['Visibility (10m)'],holdno['Visibility (10m)'])
```

```python
# m) Test Null Average Rented Bike count of seasons equal

data['Rented Bike Count'].groupby(data.Seasons).mean()
```

```python
summer = data[data.Seasons == 'Summer']
sprint = data[data.Seasons == 'Spring']
autumn = data[data.Seasons == 'Autumn']
winter = data[data.Seasons == 'Winter']
```

```python
from scipy.stats import f_oneway
```

```python
f_oneway(summer['Rented Bike Count'],sprint['Rented Bike Count'],autumn['Rented Bike Count'],winter['Rented Bike Count'])
```

```python
# n) Test Null Average Visibility of seasons equal
f_oneway(summer['Visibility (10m)'],sprint['Visibility (10m)'],autumn['Visibility (10m)'],winter['Visibility (10m)'])
```

```python
# o) Test Null No Association between seasons and holiday
from scipy.stats import chi2_contingency
```

```python
chi2_contingency(pd.crosstab(data.Seasons, data.Holiday))
```

```python
# p) Test Null No Association between holiday and functional
chi2_contingency(pd.crosstab(data.Holiday,data['Functioning Day']))
```

```python
# q) get dummies categorical data
objcols_dummy = pd.get_dummies(objcols)
```

```python
# r) Scale Numeric Data
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
numcols = numcols.drop(['Rented Bike Count'], axis = 1)
```

```python
numcols_scaled = scale.fit_transform(numcols)
```

```python
numcols_scaled = pd.DataFrame(numcols,columns=numcols.columns)
```

```python
combinedf = pd.concat([numcols_scaled,objcols_dummy],axis = 1)
```

```python
combinedf
```

```python
y = data['Rented Bike Count']
X = combinedf
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# s) Build the following Regression Models
# - Multiple Linear Regression
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

```

```python
lrmodel = lr.fit(X_train,y_train)
```

```python
# R Square
lrmodel.score(X_train,y_train)
```

```python
lrmodel.intercept_
```

```python
lrmodel.coef_
```

```python
lrmodel_pred = lrmodel.predict(X_train)
```

```python
reg_resid = y_train - lrmodel_pred
```
