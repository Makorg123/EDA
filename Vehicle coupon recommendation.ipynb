```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = data.drop('car',axis = 1)
```

```python
data = data.drop('toCoupon_GEQ5min',axis = 1)
```

```python
data[['CoffeeHouse','Restaurant20To50','CarryAway','RestaurantLessThan20','Bar']]
```

```python
for i in data.columns:
    data[i] = data[i].fillna(data[i].value_counts().idxmax())
```

```python
data.isnull().sum()
```

```python
dummy_encoded = pd.get_dummies(data,columns=['Restaurant20To50','RestaurantLessThan20','CarryAway',
                                             'CoffeeHouse','Bar','income','occupation','education',
                                             'maritalStatus','age','gender','expiration','coupon',
                                             'time','weather','passanger','destination','temperature',
                                             'direction_same','direction_opp','toCoupon_GEQ15min',
                                             'toCoupon_GEQ25min','has_children'])
```

```python
dummy_encoded
```

```python
from scipy.stats import chi2_contingency
```

```python
# Test Null no Association between Y and Gender
chi2_contingency(pd.crosstab(data.Y,data.gender))
```

```python
# Test Null no Association between Y and destination
chi2_contingency(pd.crosstab(data.Y,data.destination))
```

```python
# Test Null no Association between Y and marital status
chi2_contingency(pd.crosstab(data.Y,data.maritalStatus))
```

```python
# Test Null no Association between Y and income
chi2_contingency(pd.crosstab(data.Y,data.income))
```

```python
# Test Null no Association between Y and weather
chi2_contingency(pd.crosstab(data.Y,data.weather))
```

```python
X = dummy_encoded.drop('Y',axis = 1)
y = dummy_encoded.Y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y = LabelEncoder().fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
pd.DataFrame(y_train).value_counts()
```

```python
pd.DataFrame(y_test).value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logit= LogisticRegression()
```

```python
logitmodel = logit.fit(X,y)
```

```python
logitmodel.score(X,y)
```

```python
logitpredict = logitmodel.predict(X)
```

```python
pd.crosstab(y,logitpredict)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y,logitpredict))
```
