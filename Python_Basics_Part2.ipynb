```python
# WAF called "max_of_3Numbers" using def and if statement. The function must identified the largest number of 3 input parameters
```

```python
def max_of_3Number():
    n1,n2,n3 = int(input("Enter Num1:")),int(input("Enter Num2:")),int(input("Enter Num3:"))
    if n1>n2 and n1>n3:
        print('Num1 is Max')
    elif n2>n1 and n2>n3:
        print("Num2 is Max")
    elif n3>n1 and n3>n2:
        print("Num3 is Max")
    else:
        print("All three are equal")
```

```python
max_of_3Number()
```

```python
def max_4Number():
    n1,n2,n3,n4 = int(input("Enter Num1:")),int(input("Enter Num2:")),int(input("Enter Num3:")),int(input("Enter Num4: "))
    if n1 >n2 and n1>n3 and n1>n4:
        print("Num1 is Max")
    elif n2>n1 and n2>n3 and n2>n4:
        print("Num2 is Max")
    elif n3>n1 and n3>2 and n3>n4:
        print("Num3 is Max")
    elif n4>n1 and n4>2 and n4>n3:
        print("Num4 is Max")
    else:
        print("All are equal")
```

```python
# Find a factorial of a given number

num = int(input("Enter a number: "))
factorial = 1
if num < 0:
    print("No Factorial for Negative Numbers")
elif num == 0:
    print("Factorial of 0 is 1")
else:
    for i in range(1,num+1):
        factorial = factorial*i
    print("The factorial of",num,"is",factorial)
```

```python
# lamdba functions or anonymous functions. Functions without name. lambda keyword is must before aurgument or statement.
```

```python
# Find even numbers and odd numbers from a given range of numbers

start = int(input("Enter a number: "))
end = int(input("Enter a number: "))

even_numbers = list(filter(lambda x:x%2==0,range(start,end+1)))
print("Even Numbers are:",even_numbers)

odd_numbers  = list(filter(lambda x:x%2!=0,range(start,end+1)))
print("Odd Numbers are:",odd_numbers)
```

```python
words = str(input("Enter a sentance: "))
words = list(words.split())
print(words)

results = map(lambda w:[w.lower(),w.upper(),len(w)],words)
for i in results:
    print(i)
```

```python
start = int(input("enter a Number: "))
end   = int(input("enter a number: "))
num_list = list(range(start,end+1,1))
def cube(n):
    return (n**3)
def square(n):
    return (n**2)
funcs = [cube,square]

for i in num_list:
    results = map(lambda x:x(i),funcs)
    print(list(results))
```
