Hypothesis Testing

```python
# Test Null Average balance of y is equal

data.balance.groupby(data.y).mean()
```

```python
yyes = data[data.y == 'yes']
yno  = data[data.y == 'no']
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(yyes.balance, yno.balance)
```

```python
# Test Null Average balance of housing is equal

data.balance.groupby(data.housing).mean()
```

```python
hyes = data[data.housing == 'yes']
hno  = data[data.housing == 'no']
```

```python
ttest_ind(hyes.balance, hno.balance)
```

```python
# Test Null Average balance of marital is equal
data.balance.groupby(data.marital).mean()
```

```python
mardiv = data[data.marital == 'divorced']
marmar = data[data.marital == 'married']
marsin = data[data.marital == 'single']
```

```python
from scipy.stats import f_oneway
```

```python
f_oneway(mardiv.balance, marmar.balance, marsin.balance)
```

```python
# Test Null Average duration of education is equal
data.duration.groupby(data.education).mean()
```

```python
edpri = data[data.education == 'primary']
edsec = data[data.education == 'secondary']
edter = data[data.education == 'tertiary']
edunk = data[data.education == 'unknown']
```

```python
f_oneway(edpri.duration, edsec.duration, edter.duration, edunk.duration)
```

```python
mall = pd.get_dummies(mall, columns = ['Genre'])
```

```python
mall
```

```python
from sklearn.cluster import KMeans
```

```python
wccs_list=[]
for i in range(1,20):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(mall)
    wccs_list.append(kmeans.inertia_)
```

```python
plt.plot(range(1,20),wccs_list)
plt.title("Elbow Plot")
plt.xlabel("Cluster")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
cluster8model = KMeans(n_clusters=8)
```

```python
cluster8model = cluster8model.fit(mall)
```

```python
silhouette_score(mall,cluster8model.labels_)
```

```python
plt.scatter(mall['Annual_Income_(k$)'],mall['Spending_Score'], c=cluster8model.labels_,cmap='plasma')
```
